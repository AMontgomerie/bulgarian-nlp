```
from models.textannotation import TextAnnotator

bgnlp = TextAnnotator()
bgnlp.get_device()
```

```
input = 'България е член на ЕС.'
annotations = bgnlp(input)
print(input, '\n')
print('POS:')
for token in annotations['tokens']:
    print("{} : {}".format(token['text'].ljust(10), token['pos']))
print()
print('NER:')
for entity in annotations['entities']:
    print("{} : {}".format(entity['text'].ljust(10), entity['type']))
```
