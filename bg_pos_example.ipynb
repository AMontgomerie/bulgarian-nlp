We have some example sentences from the bg_btb-ud-test set and their correct tags. Let's pass the sentences into the model to generate some predictions.

```
from models.postagger import POSTagger

sentences = [
    {'text': 'Да не би да съм закъснял?',
    'tags': ['AUX', 'PART', 'AUX', 'AUX', 'AUX', 'VERB', 'PUNCT']},
    {'text': 'Ти да не оглуша?',
    'tags': ['PRON', 'AUX', 'PART', 'VERB', 'PUNCT']},
    {'text':'Дощя му се под покрив да живее.',
    'tags': ['VERB', 'PRON', 'PRON', 'ADP', 'NOUN', 'AUX', 'VERB', 'PUNCT']},
    {'text': 'Донесете от кухнята чаша вода.',
    'tags': ['VERB', 'ADP', 'NOUN', 'NOUN', 'NOUN', 'PUNCT']},
    {'text': 'Писал е доста повести и драми.',
    'tags': ['VERB', 'AUX', 'ADV', 'NOUN', 'CCONJ', 'NOUN', 'PUNCT']}
]

pos_tagger = POSTagger()

for sentence in sentences:
    predicted_tags = pos_tagger.generate_tags(sentence['text'])
    print('predicted: {}'.format(predicted_tags))
    print('real:      {}'.format(sentence['tags']))
```

For more information about the POS tags, see https://universaldependencies.org/u/pos/
As you can see in this case all the predictions were correct.
